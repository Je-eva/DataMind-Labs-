```python
import pandas as pd



# Read the CSV file
df = pd.read_csv('12Months_Data.csv')

df.head()

#top->mid>leaf
```

```python
# Calculate total revenue for each SKU by simply using the 'sales' column
# Here, we're assuming that the sales column represents the revenue for each SKU.
df['total_revenue'] = df['sales']

# Group the data by 'leaf_slug' and calculate the total revenue for each SKU
# We group the data based on 'leaf_slug', which is essentially the product category.
# By summing the 'total_revenue' for each group, we get the total revenue for each SKU.
sku_revenue = df.groupby('leaf_slug')['total_revenue'].sum()

# Calculate the overall total revenue across all SKUs
# We sum up all the individual SKU revenues to get the total revenue generated by the company.
total_revenue = sku_revenue.sum()

# Sort the SKUs by revenue in descending order
# This helps us understand which SKUs are generating the most revenue.
sku_revenue = sku_revenue.sort_values(ascending=False)

# Calculate cumulative revenue to identify which SKUs contribute the most to total revenue
# We calculate the cumulative sum of the sorted revenue values. This allows us to see how the revenue accumulates
# and helps identify the SKUs that together contribute the most to total revenue.
cumulative_revenue = sku_revenue.cumsum()

# Filter out the SKUs that contribute to at least 80% of the total revenue
# This step follows the 80/20 rule (Pareto principle), where we focus on the top contributors,
# i.e., the SKUs that make up 80% of the total revenue.
top_skus = sku_revenue[cumulative_revenue <= total_revenue * 0.80]

# Convert the filtered results into a DataFrame and reset the index
# After filtering the SKUs, we convert the series back into a DataFrame and reset the index for better readability.
top_skus_df = top_skus.to_frame(name="total_revenue")
top_skus_df.reset_index(inplace=True)

# Convert the DataFrame into a list of lists, including the header, so it can be added to Excel
# We prepare the data by converting it into a list of lists (one for the header and one for the values).
data_to_add = top_skus_df.values.tolist()
header = top_skus_df.columns.tolist()  # Extract the column names

# Combine the header and the data into a single list to prepare it for export
# We combine the header with the data so it's ready to be added to an Excel sheet.
final_data = [header] + data_to_add

# Add the processed data to a new sheet in the Excel file
# Finally, we save the filtered data into a new sheet called "Top SKUs" in the Excel workbook.
add_data_to_excel(final_data, filename="DataMindLabs_JeevaVinod_Data_Analysis.xlsx", sheet_name="Top SKUs")
```

```python
# Group the data by 'leaf_slug', 'mid_slug', and 'top_slug' to calculate total quantity sold for each SKU combination
# We are grouping by three hierarchical levels (leaf_slug, mid_slug, and top_slug) to understand how different
# combinations of product categories contribute to the total quantity sold.
sku_quantity = df.groupby(['leaf_slug', 'mid_slug', 'top_slug'])['total_qty'].sum().reset_index()

# Find the highest selling combination (by total_qty) for each 'leaf_slug'
# We use the `groupby` function to find the SKU combination with the highest total quantity sold for each 'leaf_slug'.
# This gives us the most popular combination for each leaf_slug category.
highest_selling = sku_quantity.loc[sku_quantity.groupby('leaf_slug')['total_qty'].idxmax()]

# Reset the index to clean up the resulting DataFrame
# After filtering the highest-selling combinations, we reset the index to make the DataFrame look cleaner and more manageable.
highest_selling = highest_selling.reset_index(drop=True)

# Select the relevant columns: leaf_slug, mid_slug, top_slug, and total_qty
# Here, we choose only the columns that are important for our analysis: the product categories and the total quantity sold.
highest_selling_data = highest_selling[['leaf_slug', 'mid_slug', 'top_slug', 'total_qty']]

# Convert the DataFrame to a list of lists (including the header)
# We need to convert the DataFrame to a list format because we plan to add it to an Excel sheet.
data_to_add_highest_selling = highest_selling_data.values.tolist()

# Extract the header (column names) for the DataFrame to be included in the final output
header_highest_selling = highest_selling_data.columns.tolist()

# Combine header and data into a single list, so it can be added to Excel properly
final_data_highest_selling = [header_highest_selling] + data_to_add_highest_selling

# Add the processed data to Excel in a new sheet called "Highest Selling SKUs"
# Finally, we add this cleaned and organized data to an Excel sheet, making it available for reporting or further analysis.
add_data_to_excel(final_data_highest_selling, filename="DataMindLabs_JeevaVinod_Data_Analysis.xlsx", sheet_name="Highest Selling SKUs")
```

```python
import pandas as pd

# ... (Your existing code to load the DataFrame df) ...

# Calculate the % contribution of average_sp (Sales Price) to avg_bill_value (Average Bill Value) for each row.
# This will help us understand how much each SKU contributes to the overall bill amount.
df['sp_to_bill_percentage'] = (df['average_sp'] / df['avg_bill_value']) * 100

# Group the data by 'leaf_slug' (which represents the SKU) and calculate the *mean* of the % contribution.
# Grouping by SKU helps us understand how each SKU performs across different transactions or records.
sku_contribution = df.groupby('leaf_slug')['sp_to_bill_percentage'].mean().reset_index()

# Sort the SKUs in descending order of their contribution.
# We want to prioritize the SKUs with the highest percentage contribution, as they are the most important ones.
sku_contribution = sku_contribution.sort_values(by='sp_to_bill_percentage', ascending=False)

# Extract the top 10 SKUs based on their % contribution.
# This allows us to focus on the most impactful SKUs.
top_10_contributions = sku_contribution.head(10)

# Improved Display: Output the top 10 SKUs with their % contribution formatted to two decimal places.
# This makes the output easier to read and interpret.
print("Top 10 SKUs by % Contribution of Average SP to Average Bill Value:\n")
for index, row in top_10_contributions.iterrows():
    print(f"{row['leaf_slug']}: {row['sp_to_bill_percentage']:.2f}%") # Format to two decimal places for better readability

# Write the data to Excel for further analysis or reporting.
# This step ensures that the results are saved and can be shared with others or used in a report.
data_to_add = top_10_contributions.values.tolist()  # Extract data and headers
header = top_10_contributions.columns.tolist()  # Get column names as the header
final_data = [header] + data_to_add  # Combine header and data into a final structure suitable for export

# Add the data to Excel using the defined 'add_data_to_excel' function.
# This saves the results in a sheet called "Top 10 SKU Contributions" for easy access.
add_data_to_excel(final_data, filename="DataMindLabs_JeevaVinod_Data_Analysis.xlsx", sheet_name="Top 10 SKU Contributions")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is already loaded and contains the 'date' and 'sales' columns
# If 'date' is not in datetime format, uncomment the following lines to convert it:
# df['date'] = pd.to_datetime(df['date'])
# df.set_index('date', inplace=True)

# Weekly revenue calculation:
# Resample the 'sales' data to weekly frequency, summing up the sales for each week.
# This allows us to analyze trends on a weekly basis, making the data more manageable.
weekly_revenue = df['sales'].resample('W').sum()

# Calculating a 4-week rolling mean:
# A rolling mean helps smooth out fluctuations and highlights trends more clearly.
# In this case, we're using a 4-week rolling mean to understand the average revenue over a month.
rolling_mean = weekly_revenue.rolling(window=4).mean()

# Plotting the Area Plot:
# The 'fill_between' function creates a shaded area between the x-axis and the 'weekly_revenue' values.
# This gives a visual sense of the changes in revenue over time.
plt.figure(figsize=(12, 6))  # Set the figure size for better clarity
plt.fill_between(weekly_revenue.index, weekly_revenue, color='skyblue', alpha=0.4, label='Weekly Revenue')

# Plot the 4-week rolling mean on top of the area plot to provide a smooth trend line.
# The rolling mean will help mitigate the effect of short-term fluctuations.
plt.plot(rolling_mean, color='red', label='4-Week Rolling Average', linewidth=2)

# Formatting the y-axis for better readability:
# We use a custom formatter to display the y-axis without the rupee symbol, but with a comma separator.
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.FuncFormatter(lambda x, loc: f'{x:,.2f}'))

# Adding titles and labels to make the chart more informative:
plt.title('Weekly Revenue Trend with 4-Week Rolling Average (Area Plot)')
plt.xlabel('Week')
plt.ylabel('Revenue')
plt.grid(True)  # Add gridlines for better readability
plt.legend()  # Show legend to identify each line

# Save the plot as an image:
# Saving the plot as an image ensures we can share or embed it later in reports or dashboards.
plt.savefig('weekly_revenue_trend.png')

# Now, add this plot to a new sheet in the existing Excel file:
from openpyxl import load_workbook
from openpyxl.drawing.image import Image

# Load the existing Excel workbook:
wb = load_workbook('DataMindLabs_JeevaVinod_Data_Analysis.xlsx')

# Add a new sheet called 'Weekly Revenue Trend' to the workbook:
ws = wb.create_sheet('Weekly Revenue Trend')

# Insert the saved image into the new sheet, starting at cell 'A1':
img = Image('weekly_revenue_trend.png')
ws.add_image(img, 'A1')

# Save the updated workbook with the new sheet and image:
wb.save('DataMindLabs_JeevaVinod_Data_Analysis.xlsx')

# Show the plot in the output:
plt.show()
```
